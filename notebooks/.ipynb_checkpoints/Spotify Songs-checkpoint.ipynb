```python
import numpy as np
import pandas as pd
import hvplot.pandas

pd.set_option('display.max_columns', 50)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv')
```

```python
df
```

### Distribution of `playlist_genre`

```python
df['playlist_genre'].value_counts().hvplot(kind='bar')
```

### Top 10 `track_artist`

```python
df['track_artist'].value_counts().head(10).hvplot(kind='bar', rot=20)
```

### Top 5 artist by genre 

```python
df_grouped = (
    df
    .groupby(['playlist_genre', 'track_artist'])
    .agg(n = ('track_id', 'count'))
    .reset_index()
    .sort_values(['playlist_genre', 'n'], ascending=False)
    .groupby('playlist_genre')
    .head(5)
)
df_grouped
```

```python
df_grouped.hvplot(kind='bar', groupby='playlist_genre', x='track_artist', y='n')
```

```python
rock_top5 = df_grouped.query("playlist_genre == 'rock'").hvplot(kind='bar', x='track_artist', y='n', title='rock')
rap_top5  = df_grouped.query("playlist_genre == 'rap'").hvplot(kind='bar', x='track_artist', y='n', title='rap')
```
